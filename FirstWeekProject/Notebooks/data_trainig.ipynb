```python
import pandas as pd
import numpy as np

from joblib import dump
import os

from sklearn.metrics import r2_score, mean_absolute_error, root_mean_squared_error

from sklearn.model_selection import train_test_split, KFold,cross_val_score
from sklearn.feature_selection import RFE

from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import NearestNeighbors
from xgboost import XGBRegressor

Xgb_Reg = XGBRegressor()
N_neig = NearestNeighbors()
DT_Reg = DecisionTreeRegressor()
RF_Reg = RandomForestRegressor()
SV_Reg = SVC()
LR_Reg = LogisticRegression()
```

# ----------------------------------------------------------------------

# datasetni ajratamiz

```python
x = df1.drop('WIND', axis=1)
y = df1['WIND']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=99)
```

# Data ni alorithmlar bilan train qilamiz

# --------------------------------------------------------------

# XGBoost

```python
grBoost = GradientBoostingRegressor(n_estimators=199, max_depth=4, random_state=99)

Rfe = RFE(grBoost, n_features_to_select=7)
Rfe.fit(x_train, y_train)
SelectedFeatures = x_train.columns[Rfe.support_]

grBoost.fit(x_train[SelectedFeatures], y_train)
pred1 = grBoost.predict(x_test[SelectedFeatures])

print(f"R2 = {r2_score(y_test,pred1)}")
print(f"MAE = {mean_absolute_error(y_test, pred1)}")
```

# Random forest

```python
Rfe = RFE(RF_Reg, n_features_to_select=7) 
Rfe.fit(x_train, y_train)
selected_features = x_train.columns[Rfe.support_]

RF_Reg.fit(x_train[selected_features], y_train)
pred3 = RF_Reg.predict(x_test[selected_features])

print(f"R2 = {r2_score(y_test,pred3)}")
print(f"MAE = {mean_absolute_error(y_test, pred3)}")
```

# Desicion tree

```python
Rfe = RFE(DT_Reg, n_features_to_select=8)
Rfe.fit(x_train, y_train)
selected_features = x_train.columns[Rfe.support_]

DT_Reg.fit(x_train[selected_features], y_train)
pred4 = DT_Reg.predict(x_test[selected_features])

print(f"R2 = {r2_score(y_test,pred4)}")
print(f"MAE = {mean_absolute_error(y_test, pred4)}")
```

# dataset ni ajratamiz

```python
x = df.drop('WIND',axis=1)
y = df['WIND']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=99)
```

# Gradient Boosting

```python
grBoost = GradientBoostingRegressor(n_estimators=500,learning_rate=0.05, max_depth=4, random_state=99)

Rfe = RFE(grBoost, n_features_to_select=5)
Rfe.fit(x_train, y_train)
SelectedFeatures = x_train.columns[Rfe.support_]

grBoost.fit(x_train[SelectedFeatures], y_train)
pred1 = grBoost.predict(x_test[SelectedFeatures])

print(f"R2 = {r2_score(y_test,pred1)}")
print(f"MAE = {mean_absolute_error(y_test, pred1)}")
```

# XGBoost

```python
Rfe = RFE(Xgb_Reg, n_features_to_select=10)
Rfe.fit(x_train, y_train)
selected_features = x_train.columns[Rfe.support_]

Xgb_Reg.fit(x_train[selected_features], y_train)
pred2 = Xgb_Reg.predict(x_test[selected_features])

print(f"R2 = {r2_score(y_test,pred2)}")
print(f"MAE = {mean_absolute_error(y_test, pred2)}")
```

# Random forest

```python
Rfe = RFE(RF_Reg, n_features_to_select=7) 
Rfe.fit(x_train, y_train)
selected_features = x_train.columns[Rfe.support_]

RF_Reg.fit(x_train[selected_features], y_train)
pred3 = RF_Reg.predict(x_test[selected_features])

print(f"R2 = {r2_score(y_test,pred3)}")
print(f"MAE = {mean_absolute_error(y_test, pred3)}")
```

# Decision tree 

```python
Rfe = RFE(DT_Reg, n_features_to_select=5)
Rfe.fit(x_train, y_train)
selected_features = x_train.columns[Rfe.support_]

DT_Reg.fit(x_train[selected_features], y_train)
pred4 = DT_Reg.predict(x_test[selected_features])

print(f"R2 = {r2_score(y_test,pred4)}")
print(f"MAE = {mean_absolute_error(y_test, pred4)}")
```
