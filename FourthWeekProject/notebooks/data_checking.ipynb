```python
from sklearn.preprocessing import OrdinalEncoder

def ordinal_custom(df, column, order):
    encoder = OrdinalEncoder(categories=[order])
    df[[column]] = encoder.fit_transform(df[[column]])
    return df, encoder

order = ["Primary", "Secondary", "Bachelor", "Master", "PhD"]
df, enc = ordinal_custom(df, "education", order)
```

```python
columns = ['diet_quality', 'parental_education_level','internet_quality']

oreder = []

diet_oreder = ['Poor','Fair','Good']
parent_edu_order = ['High School','Bachelor', 'Master']
internet_qual_oreder = ['Poor','Average','Good']

for column in columns:
    if column == columns[0]:
        oreder = diet_oreder
    elif column == columns[1]:
        oreder = parent_edu_order
    else:
        oreder = internet_qual_oreder
        
    encoder(df=df,order=oreder,column=column)
```
