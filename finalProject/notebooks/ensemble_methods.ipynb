```python
x = df.drop(columns=['Range_km_level','Price_level'])
y = df[['Range_km_level','Price_level']]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.19,random_state=99)
```

```python
from sklearn.ensemble import (
    BaggingClassifier,
    GradientBoostingClassifier,
    VotingClassifier,
    StackingClassifier,
    RandomForestClassifier,
    ExtraTreesClassifier
)
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.multioutput import MultiOutputClassifier
from sklearn.metrics import accuracy_score
from xgboost import XGBClassifier

# BAGGING
bagging_base = BaggingClassifier(
    estimator=DecisionTreeClassifier(random_state=42),
    n_estimators=100,
    random_state=99
)
bagging_model = MultiOutputClassifier(bagging_base)


# BOOSTING (XGBoost)
boosting_base = XGBClassifier(
    learning_rate=0.05, 
    n_estimators=99, 
    max_depth=5, 
    random_state=42
)
boosting_model = MultiOutputClassifier(boosting_base)


# VOTING
voting_base = VotingClassifier(
    estimators=[
        ('rf', RandomForestClassifier(random_state=42)),
        ('knn', KNeighborsClassifier()),
        ('lr', LogisticRegression(max_iter=1000))
    ],
    voting='hard'
)
voting_model = MultiOutputClassifier(voting_base)


#  STACKING
stacking_base = StackingClassifier(
    estimators=[
        ('xt', ExtraTreesClassifier(random_state=42)),
        ('lr', LogisticRegression(max_iter=1000)),
        ('svm', SVC(random_state=42))
    ],
    final_estimator=RandomForestClassifier(max_depth=5, random_state=99)
)
stacking_model = MultiOutputClassifier(stacking_base)


models = {
    'Bagging': bagging_model,
    'Boosting (XGB)': boosting_model,
    'Voting': voting_model,
    'Stacking': stacking_model
}


for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    print(f"\n{'-'*30}\n{name} natijasi")
    
    
    y_test_np = np.array(y_test)
    
    for i in range(y_test_np.shape[1]):
        y_test_col = y_test_np[:, i]
        y_pred_col = y_pred[:, i]
        
        accuracy = accuracy_score(y_test_col, y_pred_col)
        print(f"  Target {i+1} (Ustun {i+1}) Accuracy: {accuracy:.3f}")
```
