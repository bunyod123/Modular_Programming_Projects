```python
import pandas as pd
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv(r"C:\Users\bunyo\OneDrive\Desktop\AI_Course\ModularProgramProjects\finalProject\data\preprocessed_data\preprocessed_data.csv")
```

```python
df.sample(2)
```

```python
df['Range_km_level'].value_counts()
```

```python
df['Price_level'].value_counts()
```

```python
df['Range_km_level'] = df['Range_km_level'].replace([0.0,0.5,1.0],[0,1,2]).astype(int)
df['Price_level'] = df['Price_level'].replace([0.0,0.5,1.0], [0,1,2]).astype(int)
df = df.drop(columns=['Range_km','Price_Euro'])
```

```python
df.head()
```

# Smote

```python
x = df.drop(columns=['Price_level'])
y = df['Price_level']
smote = SMOTE(random_state=99)


x_train, x_test, y_train,y_test = train_test_split(x, y, test_size=0.19, random_state=99)

print(f"Before sampling\n : {y_train.value_counts()}")

x_train_balanced, y_train_balanced = smote.fit_resample(x_train,y_train)
print(f"\nAfter sampling:\n {y_train_balanced.value_counts()}")
```

```python
df_train = x_train_balanced.copy()
df_train['Price_level'] = y_train_balanced

df_test = x_test.copy()
df_test['Price_level'] = y_test

df_final_price = pd.concat([df_train,df_test], axis=0, ignore_index=True)
df_final_price.info()
```
