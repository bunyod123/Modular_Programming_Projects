```python
cols = ['Range_km','Efficiency','Price']

df[cols] = df[cols].replace(r"[^ 0-9 , .]", '', regex=True)

df[['Range_km','Efficiency']] = df[['Range_km','Efficiency']].astype(int)
df['Price'] = df['Price'].str.replace(',','.').astype(float)

df = df.rename(columns={'Weight':'Weight_kg', 'Firth_stop_range':'Firth_stop_range_km','Price':'Price_Euro'})
```

```python
df1 = df[df['Price_Euro'].isnull()]

df1['Price_Euro'] = (df1['Range_km'] * df1['Price/Range(km)']) / 1000
```
