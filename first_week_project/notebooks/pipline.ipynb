```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler, OrdinalEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np
```

```python
path = r'C:\Users\bunyo\OneDrive\Desktop\AI_Course\ModularProgramProjects\first_week_project\data\raw_data\halfPreparedData.csv'
df = pd.read_csv(path)
```

```python
df.head()
```

```python
df.info()
```

```python
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
lab_cols = df.select_dtypes(exclude=[np.number]).columns.tolist()
```

```python
preprocessor = ColumnTransformer(
    [
        ('num', MinMaxScaler(),num_cols),
        ('cat', OrdinalEncoder(handle_unknown='use_encoded_value',unknown_value=-1), lab_cols)
    ]
)
```

```python
preprocessor
```

```python
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('model',RandomForestClassifier(random_state=99))
])
```

```python
pipeline
```

```python
x = df.drop("Score_class",axis=1)
y = df['Score_class']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.18,random_state=99)
```

```python
y_train
```

```python
pipeline.fit(x_train,y_train)
```
