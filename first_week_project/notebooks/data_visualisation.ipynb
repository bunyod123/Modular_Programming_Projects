```python
data_types = df.dtypes.value_counts()
plt.figure(figsize=(6,4))
plt.pie(data_types, labels=['str','float'], explode=[0.06,0.06],autopct='%1.1f%%',startangle=90,shadow=True)
plt.title("Data type values")
plt.show()
```

```python
sns.set_style('whitegrid')
sns.stripplot(x='Name', y= 'Score', data=df,jitter=True)
sns.despine()
```
