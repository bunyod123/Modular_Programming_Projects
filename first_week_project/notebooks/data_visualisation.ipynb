```python
df.boxplot(column='social_media_hours')
```

```python
q1 = df['social_media_hours'].quantile(.25)
q3 = df['social_media_hours'].quantile(.75)
IQR = q3 - q1
```

```python
lower_bound = q1 - 1.5 * IQR
upper_bound = q3 + 1.5 * IQR
```

```python
df[df['social_media_hours'] > upper_bound]
```
