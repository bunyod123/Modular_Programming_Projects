```python
import numpy as np

# Ustun nomlari aniq bo'lishi uchun o'zgaruvchilarga olamiz
# (O'zingizdagi aniq ustun nomlarini yozing, masalan 'Universitet Score' va 'Rank Score')
score_col = 'Score'
rank_col = 'Acedemik Reputation'

# 1. Ikkala ustunda ham ma'lumoti bor (NaN bo'lmagan) dastlabki 750 ta qatorni ajratib olamiz
valid_data = df.dropna(subset=[score_col, rank_col])

x_valid = valid_data[rank_col]
y_valid = valid_data[score_col]

# 2. Ikkala ustun o'rtasidagi matematik bog'liqlikni (trendni) topamiz
# 1 raqami chiziqli bog'liqlik (to'g'ri chiziq) ekanligini bildiradi
trend_modeli = np.polyfit(x_valid, y_valid, 1)
hisoblash_formulasi = np.poly1d(trend_modeli)

# 3. 'Universitet Score' da NaN bo'lgan qatorlar indeksini topamiz
nan_indekslar = df[score_col].isnull()

# 4. Shu qatorlardagi 'Rank Score' asosida yangi score'larni hisoblab, df ga joylaymiz
df.loc[nan_indekslar, score_col] = hisoblash_formulasi(df.loc[nan_indekslar, rank_col])

# Natijani tekshiramiz
print(df[[rank_col, score_col]])
```
