# Score da yoq qiymatlarni Acedemik Reputation ga asoslanib toldirish

```python
import numpy as np

score_col = 'Score'
rank_col = 'Acedemik Reputation'


valid_data = df.dropna(subset=[score_col, rank_col])

x_valid = valid_data[rank_col]
y_valid = valid_data[score_col]


trend_modeli = np.polyfit(x_valid, y_valid, 1)
hisoblash_formulasi = np.poly1d(trend_modeli)

# Score da NaN bo'lgan qatorlar
nan_indekslar = df[score_col].isnull()

df.loc[nan_indekslar, score_col] = hisoblash_formulasi(df.loc[nan_indekslar, rank_col])


print(df[[rank_col, score_col]])
```

# Universitet Score ga kora 5 qismga bolindi -> Multi-class Classification uchun

```python

chegara = [0, 35, 55, 70, 82, 100]

toifa = ['good', 'very good', 'great', 'really great', 'best']

#  kesish
df['Score_class'] = pd.cut(df['Score'], bins=chegara, labels=toifa)

df.head()
```
