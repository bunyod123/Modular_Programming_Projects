```python
import pandas as pd
import numpy as np

from joblib import dump
import os

from sklearn.metrics import classification_report, accuracy_score

from sklearn.model_selection import train_test_split, KFold,cross_val_score
from sklearn.feature_selection import RFE

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import NearestNeighbors
from xgboost import XGBClassifier

Xgb_Class = XGBClassifier()
N_neig = NearestNeighbors()
DT_Class = DecisionTreeClassifier()
RF_Reg = RandomForestClassifier()
SV_Reg = SVC()
LR_Reg = LogisticRegression()
```

# data split

```python
x = df.drop('price_range', axis=1)
y = df['price_range']
y = y.astype(int)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=99)
```

# Training

```python
grBoost = GradientBoostingClassifier(n_estimators=500,learning_rate=0.05, max_depth=4, random_state=99)

Rfe = RFE(grBoost, n_features_to_select=15)
Rfe.fit(x_train, y_train)
SelectedFeatures = x_train.columns[Rfe.support_]

grBoost.fit(x_train[SelectedFeatures], y_train)
pred1 = grBoost.predict(x_test[SelectedFeatures])

print(f"report = {classification_report(y_test,pred1)}")
```

# XGBoost

```python
Rfe = RFE(Xgb_Class, n_features_to_select=10)
Rfe.fit(x_train, y_train)
selected_features = x_train.columns[Rfe.support_]

Xgb_Class.fit(x_train[selected_features], y_train)
pred2 = Xgb_Class.predict(x_test[selected_features])

print(f"report = {classification_report(y_test,pred2)}")
```

# Random Forest

```python
Rfe = RFE(RF_Reg, n_features_to_select=11) 
Rfe.fit(x_train, y_train)
selected_features = x_train.columns[Rfe.support_]

RF_Reg.fit(x_train[selected_features], y_train)
pred3 = RF_Reg.predict(x_test[selected_features])

print(f"report = {classification_report(y_test,pred3)}")
```

# Desicion Tree

```python
Rfe = RFE(DT_Class, n_features_to_select=15)
Rfe.fit(x_train, y_train)
selected_features = x_train.columns[Rfe.support_]

DT_Class.fit(x_train[selected_features], y_train)
pred4 = DT_Class.predict(x_test[selected_features])

print(f"report = {classification_report(y_test,pred4)}")
```

# Hyperparametr tuning

```python
import optuna
import xgboost as xgb
from sklearn.model_selection import cross_val_score


def objective(trial):
    # Model turini tanlash
    classifier_name = trial.suggest_categorical("classifier", ["DecisionTree", "RandomForest", "GradientBoosting", "XGBoost"])

    if classifier_name == "DecisionTree":
        max_depth = trial.suggest_int("dt_max_depth", 2, 32, log=True)
        model = DecisionTreeClassifier(max_depth=max_depth)

    elif classifier_name == "RandomForest":
        n_estimators = trial.suggest_int("rf_n_estimators", 10, 200)
        max_depth = trial.suggest_int("rf_max_depth", 2, 32, log=True)
        model = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth)

    elif classifier_name == "GradientBoosting":
        learning_rate = trial.suggest_float("gb_learning_rate", 1e-3, 0.1, log=True)
        n_estimators = trial.suggest_int("gb_n_estimators", 50, 300)
        model = GradientBoostingClassifier(learning_rate=learning_rate, n_estimators=n_estimators)
    
    else: # XGBoost
        param = {
            "verbosity": 0,
            "objective": "binary:logistic",
            "learning_rate": trial.suggest_float("xgb_learning_rate", 1e-3, 0.1, log=True),
            "max_depth": trial.suggest_int("xgb_max_depth", 3, 10),
            "lambda": trial.suggest_float("xgb_lambda", 1e-8, 1.0, log=True),
            "alpha": trial.suggest_float("xgb_alpha", 1e-8, 1.0, log=True),
        }
        model = xgb.XGBClassifier(**param)

    # Cross-validation
    score = cross_val_score(model, x, y, n_jobs=-1, cv=3)
    accuracy = score.mean()
    return accuracy

# Optimallashtirishni boshlash
study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=50)

print(f"Eng yaxshi natija: {study.best_value}")
print(f"Eng yaxshi model va parametrlar: {study.best_params}")
```

# GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier


model = RandomForestClassifier()
param_grid = {
    'n_estimators': [10, 50, 100],
    'max_depth': [None, 5, 10],
    'criterion': ['gini', 'entropy']
}


grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=3)
grid_search.fit(x, y)

print(f"Eng yaxshi parametrlar: {grid_search.best_params_}")
print(f"Eng yaxshi natija: {grid_search.best_score_}")
```

# RandomizedSearchCV

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

param_dist = {
    'n_estimators': randint(10, 200),
    'max_depth': [None, 3, 5, 10, 20],
    'min_samples_split': randint(2, 11)
}

random_search = RandomizedSearchCV(estimator=model, param_distributions=param_dist, n_iter=10, cv=5)
random_search.fit(x, y)

print(f"Eng yaxshi parametrlar: {random_search.best_params_}")
print(f"Eng yaxshi natija: {random_search.best_score_}")
```

```python
import optuna
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

def objective(trial):
    n_estimators = trial.suggest_int('n_estimators', 10, 200)
    max_depth = trial.suggest_int('max_depth', 2, 32, log=True)
    
    model = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth)
    
    score = cross_val_score(model, x, y, cv=6).mean()
    return score

study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=20) 

print(f"Eng yaxshi parametrlar: {study.best_params}")
print(f"Eng yaxshi natija: {study.best_value}")
```
